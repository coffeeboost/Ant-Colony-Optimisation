```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./weWinThese.csv')
```

```python
data
```

```python
data['time'] = abs(data['time'] - len(data))
```

```python
data
```

```python
lines = [x for x in data.columns if x[0] == 'p']

ax = plt.gca()

for y in lines:
    data.plot(kind='line',x='time',y=y, ax=ax)

plt.savefig("everything")
plt.show()
```

```python
a = data.iloc[len(data)-1][6:]
```

```python
a = a[a>0]
a
```

```python
keep = ['time','1','2','3','4','5'] + list(a.index)
```

```python
data = data[keep]
```

```python
data
```

```python
lines = [x for x in data.columns if x[0] == 'p']

ax = plt.gca()

for y in lines:
    data.plot(kind='line',x='time',y=y, ax=ax)

plt.savefig("notEverything")
plt.show()
```
